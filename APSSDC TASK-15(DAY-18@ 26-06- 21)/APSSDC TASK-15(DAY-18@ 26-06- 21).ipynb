```python
import pandas as pd
```

```python
data=pd.read_csv("bird.csv")
data
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

## Line plot

```python
plt.plot(data["huml"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

```python
plt.plot(data["huml"][:10])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

### Histo Graph

```python
plt.hist(data["humw"],width=0.3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

```python
plt.hist(data["humw"][:10],width=0.3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

## Scatter Plot

```python
plt.scatter(data["huml"],data["humw"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

```python
plt.scatter(data["huml"][:50],data["humw"][:50])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

## Bar graph

```python
plt.bar(data["type"],data["tarl"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

## Pie chart

```python
plt.pie(data["tibl"][:10],labels=data["tarw"][:10],radius =2,autopct= "%1.2f %%")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```

## Boxplot

```python
plt.boxplot(data["huml"][:50])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.grid()
plt.show()
```
